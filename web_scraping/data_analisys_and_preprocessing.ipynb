```python
cadunico_data = cadunico[cadunico.columns[1:]].copy()
corr_matrix = cadunico_data.corr()
```

```python
threshold = .85
indices = np.where(np.abs(corr_matrix.values) > threshold)
pairs = [(corr_matrix.columns[i], corr_matrix.columns[j]) for i, j in zip(*indices) if i > j]
```
